## 4 | Labelling Categorical Data

```python
diagnosis = {
    "B": 0,
    "M": 1
}

df['diagnosis'] = df['diagnosis'].map(diagnosis)
```

Diagnosis result are categorical so they must be converted to numeric in order to use the model. Labelling the diagnosis column value with "B" = 0 and "M" = 1. Labelling is done by mapping values based on the dictionary that has been created.

```python
# Correlations between features
matrix = df.corr().round(2)
plt.figure(figsize=(16,16))
sns.heatmap(matrix, annot=True, vmax=1, vmin=-1, center=0, cmap='coolwarm')
plt.title("Features Correlation")
plt.show()
```

The correlation between numerical features can be seen in  the figure above. The correlation value is rounded to 2 numbers behind zero. The higher the value, the stronger the correlation. A positive correlation means that it is directly propotional, if the correlation is negative then the correlation is inversely propotional.

## 5 | Balancing Data + Splitting the dataset

```python
# Defining x and y
x = df.drop(columns=['diagnosis'])
y = df['diagnosis']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
from imblearn.over_sampling import SMOTE
SMOTE = SMOTE()
x_train, y_train = SMOTE.fit_resample(x_train, y_train)
```

```python
x_train.shape, x_test.shape
```

## 5 | Modelling

### 5.1 Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingClassifier
gb = GradientBoostingClassifier()
```

```python
# Training the model
gb.fit(x_train, y_train)
```

```python
# Predict testing test
y_pred_gb = gb.predict(x_test)
```

```python
print('Training-set accuracy score: ', gb.score(x_train, y_train))
print('Test-set accuracy score: ', gb.score(x_test, y_test))
```

```python
plt.figure(figsize=(6,6))
sns.heatmap(confusion_matrix(y_test, y_pred_gb), annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predict')
plt.ylabel('Actual')
plt.title('Confusion Matrix for Gradient Boosting')
plt.show()
```

```python
# Check model performance using classification_report
print(classification_report(y_test, y_pred_gb))
```

```python
cm = confusion_matrix(y_test, y_pred_gb)

specificity = cm[1,1]/(cm[1,0] + cm[1,1])
print('Specificity : ', specificity)
```

```python
# Check model performance using auc score
roc_auc_score(y_test, y_pred_gb) * 100
```

### 5.2 Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
classifier_rf = RandomForestClassifier()
```

```python
# Training the model
classifier_rf.fit(x_train, y_train)
```

```python
# Predict testing test
y_pred_rf = classifier_rf.predict(x_test)
```

```python
print('Training-set accuracy score: ', classifier_rf.score(x_train, y_train))
print('Test-set accuracy score: ', classifier_rf.score(x_test, y_test))
```

```python
plt.figure(figsize=(6,6))
sns.heatmap(confusion_matrix(y_test, y_pred_rf), annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predict')
plt.ylabel('Actual')
plt.title('Confusion Matrix for Random Forest')
plt.show()
```

```python
# Check model performance using classification_report
print(classification_report(y_test, y_pred_rf))
```

```python
cm = confusion_matrix(y_test, y_pred_rf)

specificity = cm[1,1]/(cm[1,0] + cm[1,1])
print('Specificity : ', specificity)
```

```python
# Check model performance using auc score
roc_auc_score(y_test, y_pred_rf) * 100
```
